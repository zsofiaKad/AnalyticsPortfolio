```python
#Lista de columnas a eliminar
columnas_a_eliminar = [
    'md5', 'neighbourhood_name', 'neighborhood_overview', 'neighbourhood_district',
    'availability_30', 'availability_60', 'availability_90',
    'availability_365', 'first_review_date', 'last_review_date',
    'license', 'is_instant_bookable', 'reviews_per_month'
]

df = df.drop(columns=columnas_a_eliminar)
```

```python
df.head(2)
```

# Limpieza de datos

```python
#Columna ID´s
df.rename(columns={'apartment_id': 'ID'}, inplace=True)

df["ID"] = df["ID"].astype(str)
df['ID'] = df['ID'].astype(str).str.zfill(18)
```

```python
#Cambiar las columnas de bathrooms, bedrooms y beds a int + rellenar valores nulos con 0
df['bathrooms'] = df['bathrooms'].fillna(0).astype(int)
df['bedrooms'] = df['bedrooms'].fillna(0).astype(int)
df['beds'] = df['beds'].fillna(0).astype(int)
```

```python
df.head(2)
```

Creamos df1 y df 2 para mayor visibilidad, más adelante volveremos a juntar en un unico df

```python
df1 = df[[
    'ID', 'host_id', 'minimum_nights', 'maximum_nights',
    'review_scores_accuracy', 'review_scores_cleanliness',
    'review_scores_checkin', 'review_scores_communication',
    'review_scores_location', 'review_scores_value'
]]

df2 = df[[
    'ID', 'name', 'description', 'latitude', 'longitude',
    'room_type', 'accommodates', 'bathrooms', 'beds', 'amenities_list',
    'price', 'has_availability', 'number_of_reviews', 'review_scores_rating',
    'country', 'city', 'insert_date'
]]
```

Limpieza de datos df1

```python
df1.head(5)
```

```python
#Cambar a int la columna maximum_nights
def convertir_a_entero(valor):
    if pd.isna(valor):
        return valor
    else:
        return int(float(valor))

#Aplicar la función
df1['maximum_nights'] = df1['maximum_nights'].apply(convertir_a_entero).astype('Int64')
```

```python
df1.dtypes
```

Limpieza de datos df2

```python
df2.head(3)
```

```python
#Cambiamos nombres a español para luego poder juntar con los df de AirBnb
valores_a_cambiar = {
    'Entire home/apt': 'Alojamiento entero',
    'Private room': 'Habitación privada',
    'Hotel room': 'Hotel',
    'Shared room': 'Habitación compartida'
}

# eemplazar los valores en la columna 'room_type'
df2['room_type'] = df2['room_type'].replace(valores_a_cambiar)
```

```python
df2['room_type'].unique()
```

```python
#Cambiamos la columna "price" a integer
def convertir_a_entero(valor):
    if pd.isna(valor):
        return valor
    else:
        return int(float(valor))

#Aplicar la función
df2['price'] = df2['price'].apply(convertir_a_entero).astype('Int64')
```

```python
#Cambiamos los valores de "counrty" y "city" para que comiencen con mayúscula
df2['country'] = df2['country'].str.title()
df2['city'] = df2['city'].str.title()
```

```python
#Cambiamos la columna "insert date" a tipo fecha
df2['insert_date'] = pd.to_datetime(df2['insert_date'])
```

```python
df2.dtypes
```

Unir df1 y df2

```python
df_alojamientos = pd.merge(df1, df2, on='ID')
```

```python
#Movemos las columnas para mejor visualización
df_alojamientos = df_alojamientos[['ID', 'host_id', 'name', 'description', 'price', 'has_availability', 'room_type', 'accommodates', 'bathrooms', 'beds', 'minimum_nights', 'maximum_nights', 'number_of_reviews', 'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value', 'country', 'city', 'latitude', 'longitude', 'insert_date', 'amenities_list']]
```
