## ¿Cómo ha impactado en la rentabilidad del negocio?

Queremos comparar el beneficio por unidad de producto contra las ventas por unidad de producto (esto es la rentabilidad de cada venta). Y queremos ver cómo de diferente es la rentabilidad en función si un descuento fue aplicado o no.

Preparación del Dato:

<font color='#3EA294'> ***Respuesta:*** *Anteriormente ya creamos las columnas Profit/Quantity, Sales/Quanity, y Flag_discount*

```python
#Preparar los datos
df['Discount'] = df['Flag_discount'] == 1

#Analizar la rentabilidad: agrupar los datos por 'Discount' y calcular la media de rentabilidad
profitability_analysis = df.groupby('Discount')[['Profit/Quantity', 'Sales/Quantity']].mean().reset_index()

# Visualizar los resultados
fig = px.bar(profitability_analysis, x='Discount', y=['Profit/Quantity', 'Sales/Quantity'],
             title='<b>Comparación de Rentabilidad con y sin Descuento</b>',
             barmode='group',
             color_discrete_sequence=["#35C7AB", "#F9B837"])

# Añadir detalles al gráfico
fig.update_layout(width=800,
                  height=600,
                  title_font=dict(color='darkgreen', size=18),
                  xaxis_title='Descuento Aplicado',
                  yaxis_title='Promedio por Unidad',
                  legend_title='Medida')
fig.update_xaxes(showgrid=True)
fig.update_yaxes(showgrid=True)

fig.show()
```

<font color='#3EA294'> ***Respuesta:*** *Los descuentos han tenido un impacto significativo en la rentabilidad del negocio. Aunque han aumentado las ventas por unidad, el beneficio ha disminuido considerablemente e incluso se ha vuelto negativo. Esto significa que, aunque los descuentos pueden atraer a más clientes, también afectan a la rentabilidad. Además, añado otro gráfico para confirmar mi conclusión. Se observa claramente que todos los productos generan beneficios, aunque con los productos con descuento, la empresa genera beneficios negativos.*

```python
#Preparar los datos
df['Discount'] = df['Flag_discount'] == 1

#Visualizar los resultados con un scatter plot
fig = px.scatter(df, x='Profit/Quantity', y='Sales/Quantity', color='Discount',
                 title='<b>Comparación de Rentabilidad con y sin Descuento</b>')

#Añadir detalles al gráfico
fig.update_layout(xaxis_title='Beneficio por Unidad',
                  yaxis_title='Ventas por Unidad',
                  legend_title='Descuento Aplicado')
fig.update_xaxes(showgrid=True)
fig.update_yaxes(showgrid=True)

fig.show()
```

Preguntas:

- ¿Son rentables todos los modos de envio?

- ¿Observas algún outlier? ¿Varia el resultado?

```python
#Preparar los datos
df['Discount'] = df['Flag_discount'] == 1

#Analizar la rentabilidad por modo de envío
profitability_by_ship_mode = df.groupby(['Discount', 'Ship Mode'])[['Profit/Quantity']].mean().reset_index()

#Visualizar los resultados con un boxplot
fig = px.box(profitability_by_ship_mode, x='Discount', y='Profit/Quantity', color='Ship Mode',
             title='<b>Comparación de Rentabilidad con y sin Descuento por Modo de Envío</b>',
             labels={'Discount': 'Descuento Aplicado', 'Profit/Quantity': 'Promedio por Unidad'},
             category_orders={'Discount': ['Con Descuento', 'Sin Descuento']},
             hover_name='Ship Mode',
             hover_data={'Profit/Quantity': True},
             )

#Añadir detalles al gráfico
fig.update_layout(xaxis_title='Descuento Aplicado',
                  yaxis_title='Rentabilidad por Unidad',
                  legend_title='Modo de Envío',
                  title_font=dict(size=20))

fig.update_xaxes(showgrid=True)
fig.update_yaxes(showgrid=True)

fig.show()
```

<font color='#3EA294'> ***Respuesta:*** *Se ve claramente que los productos con descuento tienen muy poco beneficio o incluso tienen un profit negativo. Aún así, me gustaría visualizarlo en otro gráfico.*

```python
import plotly.graph_objects as go

#Preparar los datos
df['Discount'] = df['Flag_discount'] == 1
profitability_by_ship_mode = df.groupby(['Discount', 'Ship Mode'])[['Profit/Quantity']].mean().reset_index()

#Visualizar los resultados con un gráfico de barras agrupadas
fig = go.Figure()

#Colores por defecto para Discount True/False
colors = {True: 'orange', False: '#3CA3D7'}

#Para manejar la leyenda de valores negativos
negative_added = False

#Agregar las barras al gráfico
for disc in [True, False]:
    for mode in profitability_by_ship_mode['Ship Mode'].unique():
        filtered_df = profitability_by_ship_mode[(profitability_by_ship_mode['Ship Mode'] == mode) & (profitability_by_ship_mode['Discount'] == disc)]
        fig.add_trace(go.Bar(
            x=filtered_df['Ship Mode'],
            y=filtered_df['Profit/Quantity'],
            name=f'Discount: {disc}',
            marker_color=[colors[disc] if x >= 0 else 'red' for x in filtered_df['Profit/Quantity']],
            width=[0.4],  # Ancho de las barras
            showlegend=mode == profitability_by_ship_mode['Ship Mode'].unique()[0]  # Mostrar la leyenda solo una vez por descuento
        ))

        #Añadir leyenda para negativos
        if not negative_added and any(x < 0 for x in filtered_df['Profit/Quantity']):
            fig.add_trace(go.Bar(
                x=[None],  # No visible data
                y=[None],
                marker_color='red',
                name='Discount: True = Negative Profit',
                showlegend=True
            ))
            negative_added = True

#Añadir detalles al gráfico
fig.update_layout(
    title='<b>Comparación de Rentabilidad con y sin Descuento por Modo de Envío</b>',
    xaxis_title='Modo de Envío',
    yaxis_title='Rentabilidad por Unidad',
    legend_title='Descuento Aplicado',
    barmode='group',
    xaxis_showgrid=True,
    yaxis_showgrid=True,
    title_font=dict(color='darkgreen', size=18),
    width=800,
    height=600
)

fig.show()
```

<font color='#3EA294'> ***Respuesta:*** *Mis conclusiones son que los descuentos pueden afectar significativamente la rentabilidad en diferentes modos de envío. Es importante considerar cuidadosamente la estrategia de descuentos para mantener un equilibrio entre ventas y beneficios. De los gráficos, podemos ver que no todos los modos de envío son igualmente rentables. Por ejemplo, el envío de “Standard Class” con descuento muestra una rentabilidad negativa. También se puede observar algún outlier: el modo de envío “Standard Class” con descuento es notablemente no rentable en comparación con los otros modos y condiciones. Además, podemos observar que sin descuentos, todos los modos de envío son rentables. El modo de envío más rentable parece ser “First Class”.*

## ¿Influye el Modo de Envios en la tasa de Devolución?

En el excel de tenemos conocimiento sobre devoluciones. Puedes "unirlo" (merge) a nuestro dataset para ver si la elección del metodo de envio puede repercutir en las devoluciones.

Preparación del dato:

```python
df_returns = pd.read_excel("SuperStore.xls", sheet_name="Returns")
```

```python
df_returns.head(3)
```

```python
#Unir los datos
df_merged = pd.merge(df, df_returns, on='Order ID', how='left')

#Calcular la tasa de devolución por modo de envío como porcentaje
return_rate_by_ship_mode = df_merged.groupby('Ship Mode')['Returned'].apply(lambda x: (x == 'Yes').mean() * 100).reset_index()
return_rate_by_ship_mode = return_rate_by_ship_mode.rename(columns={'Returned': 'Return Rate'})

#Definición de los colores
colors = {'Same Day': '#FC68B5', 'First Class': '#F4DF57', 'Second Class': '#4ECF85', 'Standard Class': '#4E9DCF'}

#Gráfico de barra
fig = px.bar(return_rate_by_ship_mode, x='Ship Mode', y='Return Rate',
             title='Tasa de Devolución por Modo de Envío',
             labels={'Ship Mode': 'Modo de Envío', 'Return Rate': 'Tasa de Devolución (%)'},
             color='Ship Mode',
             color_discrete_map=colors,
             width=800
             )

#Cambiar el color del título
fig.update_layout(title_font=dict(color='darkgreen', size=20))
fig.update_layout(bargap=0.4)

fig.show()
```

<font color='#3EA294'> ***Respuesta:*** *Podemos observar que el modo de envío sí, influye en la tasa de devolución. El modo de envío de "Same Day", que ofrece entrega el mismo día, tiene la mayor porcentaje de devoluciones con 11,7%. El modo de envío "First Day" tiene una tasa de devolución de un poco menos de 10%, mientras que los otros dos modos de envíos están entre el 6% y el 8%. Estos resultados sugieren que el modo de envío podría influir en los devoluciones de productos. En resumen, considerar el modo de envío es importante para comprender y abordar las tasas de devoluciónes, lo que puede ayudar a mejorar la satisfacción del cliente y reducir los costes asociados.*

```python
#Unir los datos
df_merged = pd.merge(df, df_returns, on='Order ID', how='left')

#Calcular la tasa de devolución por modo de envío como porcentaje
return_rate_by_ship_mode = df_merged.groupby(['Year', 'Ship Mode'])['Returned'].apply(lambda x: (x == 'Yes').mean() * 100).reset_index()
return_rate_by_ship_mode = return_rate_by_ship_mode.rename(columns={'Returned': 'Return Rate'})

#Definición de los colores
colors = {'Same Day': '#FC68B5', 'First Class': '#F4DF57', 'Second Class': '#4ECF85', 'Standard Class': '#4E9DCF'}

#Gráfico de línea
fig = px.line(return_rate_by_ship_mode, x='Year', y='Return Rate',
              title='Tasa de Devolución por Año y Modo de Envío',
              labels={'Year': 'Año', 'Return Rate': 'Tasa de Devolución (%)', 'Ship Mode': 'Modo de Envío'},
              color='Ship Mode',
              color_discrete_map=colors,
              width=800,
              markers=True,
              )

#Cambiar el color del título
fig.update_layout(title_font=dict(color='darkgreen', size=20))
fig.update_layout(bargap=0.4)

fig.show()
```

<font color='#3EA294'> ***Respuesta:*** *Al analizar las tasas de devolución por año y por tipo de envío, podemos observar que la entrega el mismo día ha experimentado una disminución significativa en la tasa de devolución. Esto podría indicar una mejora en la logística y/o la calidad del servicio de este tipo de envío. Los envíos de "First Class" y "Standard Class" muestran una tendencia bastante estable con una ligera fluctuación. Además, el modo de envío estándar parece estar muy estable y con relativamente pocas devoluciones.*
